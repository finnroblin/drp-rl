```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
def plot_normal_rv(mu, sigma, color="blue"):
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, stats.norm.pdf(x, mu, sigma),color=color,label=f"mu: {mu}, sigma: {sigma}")
    plt.legend()
    plt.show()
```

```python
plot_normal_rv(-10, 10, "red")
```

```python
plot_normal_rv(0,1,"green")
```

```python
plot_normal_rv(-2,4,"blue")
```

```python
plot_normal_rv(5,3, "orange")
```
